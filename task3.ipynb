# Task 3: Part 2 – Titanic Dataset

## 🎯 Objective

```python
from google.colab import files
import pandas as pd

# Upload file manually
uploaded = files.upload()

# Load into pandas using the real file name
titanic = pd.read_csv("Titanic-Dataset.csv")

# Quick check: show first 5 rows
titanic.head()
```

## Part 2: Titanic Dataset – Step 1 (Exploration)

We begin by exploring the dataset to understand its structure and quality:
- `.head()` → preview first rows
- `.info()` → check column types and missing values
- `.describe()` → summary statistics of numerical features

This gives us an overview of passenger details, survival status, and highlights missing values (e.g., Age, Cabin, Embarked).


```python
# Show first 5 rows
titanic.head()

# Show info about columns, datatypes, and missing values
titanic.info()

# Show summary statistics for numerical columns
titanic.describe()
```

## Part 2: Titanic Dataset – Step 2 (Data Cleaning)

We clean the dataset to handle missing values:
- **Age** → replaced with the median value  
- **Embarked** → replaced with the mode (most frequent port)  
- **Cabin** → dropped due to too many missing values  
- **Duplicates** → removed if present  

This ensures the dataset is consistent and ready for analysis in the next step.


```python
# Handle missing values

# Fill missing Age values with the median
titanic["Age"].fillna(titanic["Age"].median(), inplace=True)

# Fill missing Embarked values with the mode (most frequent value)
titanic["Embarked"].fillna(titanic["Embarked"].mode()[0], inplace=True)

# Drop the Cabin column (too many missing values)
titanic.drop(columns=["Cabin"], inplace=True)

# Remove duplicate rows if any
titanic.drop_duplicates(inplace=True)

# Quick check after cleaning
titanic.info()
```

## Part 2: Titanic Dataset – Step 3 (Data Analysis)

We analyze survival patterns using `groupby()`:
- **By Gender** → compare male vs female survival rates  
- **By Class** → compare survival rates across passenger classes (1st, 2nd, 3rd)  
- **Average Age per Class** → understand age distribution by class  
- **By Age Group** → categorize passengers into Child, Teen, Adult, Middle-Aged, Senior and compute survival rates  

This reveals key demographic and socio-economic factors influencing survival.


```python
# Survival rate by gender
print("Survival rate by gender:")
print(titanic.groupby("Sex")["Survived"].mean())

# Survival rate by class
print("\nSurvival rate by class:")
print(titanic.groupby("Pclass")["Survived"].mean())

# Average age per class
print("\nAverage age per class:")
print(titanic.groupby("Pclass")["Age"].mean())

# Survival rate by age group
titanic["AgeGroup"] = pd.cut(
    titanic["Age"],
    bins=[0, 12, 18, 40, 60, 80],
    labels=["Child", "Teen", "Adult", "Middle-Aged", "Senior"]
)

print("\nSurvival rate by age group:")
print(titanic.groupby("AgeGroup")["Survived"].mean())
```

## Part 2: Titanic Dataset – Step 4 (Filtering)

We filter the dataset to highlight specific survival groups:
- **Female passengers who survived**  
- **Children (under 12 years) who survived**  
- **First-class passengers who survived**  

This allows us to focus on priority groups and socio-economic factors that influenced survival.


```python
# Female passengers who survived
females_survived = titanic[(titanic["Sex"] == "female") & (titanic["Survived"] == 1)]
print("Female passengers who survived:")
print(females_survived.head())

# Children (Age < 12) who survived
children_survived = titanic[(titanic["Age"] < 12) & (titanic["Survived"] == 1)]
print("\nChildren who survived:")
print(children_survived.head())

# First-class passengers who survived
first_class_survived = titanic[(titanic["Pclass"] == 1) & (titanic["Survived"] == 1)]
print("\nFirst-class passengers who survived:")
print(first_class_survived.head())
```

## Part 2: Titanic Dataset – Step 5 (Insights)

Answering the key questions:

- **Who was more likely to survive? (Gender)**  
  Women had a much higher survival rate than men.  

- **Did class affect survival?**  
  Yes. First-class passengers had the highest survival rate, followed by second class, while third class had the lowest.  

- **Were children prioritized?**  
  Yes. Children (under 12) had better survival chances compared to adults, reflecting rescue priorities.  

- **What combination had the highest survival rate?**  
  Female passengers in **1st class** had the highest survival probability overall.
